```python
# Task #3: Plot pie charts
values = [20, 55, 5, 17, 3]
colors = ['g', 'r', 'y', 'b', 'm'] 
labels = ['AAPL', 'GOOG', 'T', 'TSLA', 'AMZN']
plt.figure(figsize = (10, 10))
explode = [0, 0.2, 0, 0, 0.2]
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.savefig('Charts/Stock_Portfolio1.png', dpi=300, bbox_inches='tight')
plt.title('Stock Portfolio')
```

```python
values = [20, 20, 20, 20, 20]
colors = ['g', 'r', 'y', 'b', 'm'] 
labels = ['AAPL', 'GOOG', 'T', 'TSLA', 'AMZN']
plt.figure(figsize = (10, 10))

explode = [0, 0.2, 0, 0, 0.2]
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('Stock Portfolio')
plt.savefig('Charts/Stock_Portfolio2.png', dpi=300, bbox_inches='tight')
```

```python
# Task #4: Plot Histograms
mu = df2['FB'].mean #dispersion from the mean
sigma = df2['FB'].std() # standard deviation

num_bins = 40
plt.figure(figsize = (15, 9))

plt.hist(df2['FB'], num_bins, facecolor = 'blue'); #semicolon(;) removes all the text outputs
plt.grid()



plt.title('Histogram: mu = ' + str(mu) + ', sigma = ' + str(sigma));


```

```python
# Task #8: Plot Scatterplot & Count plot (Seaborn)
import seaborn as sns
```

```python
# Load Data
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
# Create a dataFrame named df_cancer with input/output data
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))

```

```python
df_cancer
```

```python
df_cancer.head(10)
```

```python
df_cancer.tail(7) #568 reords and 31 columns
```

```python
sns.scatterplot(x= 'mean area', y= 'mean smoothness', hue = 'target', data = df_cancer)
```

```python
# Task #8: Count plot (Seaborn)
plt.figure(figsize=(5,5))
# sns.countplot(df_cancer['target'], label = 'count')
sns.countplot(x='target', data=df_cancer)
plt.title('Distribution of Target Classes')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()
```

```python
df_cancer['target'].value_counts()
```

```python
# Task #8: Plot Scatterplot
sns.scatterplot(x = 'mean radius', y = 'mean area', hue = 'target', data = df_cancer)
```

```python
# Task #9: Plot Pairplot, Distplot, and Heatmaps (Seaborn)

sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'])

```

```python
plt.figure(figsize = (30, 30))
sns.heatmap(df_cancer.corr(), annot = True)
```

```python
sns.distplot(df_cancer['mean radius'], bins = 30, color = 'b')
```

```python
class_0_df = df_cancer[ df_cancer['target'] == 0 ]
class_1_df = df_cancer[ df_cancer['target'] == 1 ]


```
