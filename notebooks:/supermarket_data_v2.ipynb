```python
#Import libraries
#Load the dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data:/supermarket_sales.csv")
df.head()
```

```python
# Inspect the structure
df.info()
df.describe(include="all")
```

```python
# Identify missing values
df.isna().sum()
```

```python
# Clean missing values
# Fill missing Product and Gender lines with "Unknown" or drop 
#df["Product line"] = df["Product line"].fillna("Unknown")
#df["Customer type"] = df["Customer type"].fillna("Unknown")
#OR

df = df.fillna("Unknown") # It applies to all null values

#OR
# Drop rows with too many missing values
# df = df.dropna(subset=["Product line", "Gender"])
```

```python
df.isna().sum()
```

```python
import pandas as pd

# If possible, start from the original df (right after reading the CSV)
# so 'Date' and 'Time' are still raw strings.

df["Date"] = pd.to_datetime(df["Date"].astype(str), errors="coerce")
df["Time"] = pd.to_datetime(df["Time"].astype(str), errors="coerce")

# Extract time-based features
df["Hour"] = df["Time"].dt.hour
df["Day"] = df["Date"].dt.day_name()
df["Month"] = df["Date"].dt.month_name()

hourly_sales = df.groupby("Hour")["Total"].sum().sort_values(ascending=False)
print("\nSales by Hour (sorted by Total Sales):\n", hourly_sales)
```

```python
# Check for duplicates
df.duplicated().sum()


```

```python
#If duplicates exist: drop them
df = df.drop_duplicates()
```

```python
# Confirm if duplicates are gone
df.duplicated().sum()
```

```python
# Validate numeric columns
# Some columns may be read as strings. Ensure they are numeric:
numeric_cols = ["Unit price", "Quantity", "Tax 5%", "Total", "cogs", 
                "gross margin percentage", "gross income", "Rating"]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
```

```python
# Final sanity check
df.info()
df.head()
```

```python
# Import the WordCloud class from the wordcloud library 
import numpy as np
from wordcloud import WordCloud

#  Define a patch method to convert a WordCloud object to a NumPy array without copying 
# This helps NumPy-based tools like ydata_profiling handle WordCloud objects correctly

def _wc_to_array_no_copy(self, copy=None):
    return np.asarray(self.to_image())

#  Apply the patch to WordCloud's internal methods 
# This overrides the default behavior so NumPy can convert WordCloud objects directly

WordCloud._to_array = _wc_to_array_no_copy
WordCloud.__array__ = _wc_to_array_no_copy
```
