```python

# Task #2: Plot Scatterplot
x = df2['FB']

y = df2['TWTR']

plt.figure(figsize = (15, 10))
plt.grid()
plt.scatter(x, y)
# Calculate and plot the trend line (linear regression) 
z = np.polyfit(x, y, 1) # fit a 1st-degree polynomial (line) 
p = np.poly1d(z) # create a polynomial function from the fit 
plt.plot(x, p(x), color='red', linewidth=2, label='Trend Line') # draw the line
plt.title('Scatterplot for Facebook vs Twitter Stock Prices', fontsize=20, fontweight='bold', color='darkblue')
plt.savefig('Charts/Scatterplot for Facebook vs Twitter Stock Prices.png', dpi=300, bbox_inches='tight')
```

```python
# Task #2: Plot Scatterplot
x = df2['FB']

y = df2['NFLX']

plt.figure(figsize = (15, 10))
plt.grid()
plt.scatter(x, y)
# Calculate and plot the trend line (linear regression) 
z = np.polyfit(x, y, 1) # fit a 1st-degree polynomial (line) 
p = np.poly1d(z) # create a polynomial function from the fit 
plt.plot(x, p(x), color='red', linewidth=2, label='Trend Line') # draw the line

plt.title('Scatterplot for Facebook vs Netflix Stock Prices', fontsize=20, fontweight='bold', color='darkblue')
plt.savefig('Charts/Scatterplot for Facebook vs Netflix Stock Prices.png', dpi=300, bbox_inches='tight')
```

```python
# Task #3: Plot pie charts
values = [20, 55, 5, 17, 3]
colors = ['g', 'r', 'y', 'b', 'm'] 
labels = ['AAPL', 'GOOG', 'T', 'TSLA', 'AMZN']
plt.figure(figsize = (10, 10))
explode = [0, 0.2, 0, 0, 0.2]
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.savefig('Charts/Stock_Portfolio1.png', dpi=300, bbox_inches='tight')
plt.title('Stock Portfolio', fontsize=20, fontweight='bold', color='darkblue')
plt.savefig('Charts/Stock Portfolio.png', dpi=300, bbox_inches='tight')
```

```python
# Task #4: Plot Histograms

# Calculate mean and standard deviation of Facebook stock prices
mu = df2['FB'].mean()     # average value of the stock prices
sigma = df2['FB'].std()   # how spread out the prices are

# Set number of bins for the histogram
#num_bins = 100            # more bins = smoother histogram

# Create a large figure for better visibility
plt.figure(figsize=(15, 9))  # width=15, height=9 inches

# Plot histogram with KDE (distribution curve)
sns.histplot(df2['FB'], bins= 100, kde=True, color='b', edgecolor='grey')
# kde=True adds the smooth curve; edgecolor outlines each bar

# Add vertical line at the mean value
plt.axvline(mu, color='red', linestyle='--', linewidth=2, label=f'Mean: {mu:.2f}')
# red dashed line shows where the average price is

# Add grid for easier reading
plt.grid()

# Add title with mean and standard deviation info
plt.title(f'Facebook Stock Price Histogram: mu = {mu:.2f}, sigma = {sigma:.2f}', fontsize=16)

# Add main title in bold blue
#plt.title('Facebook Stock Distribution', fontsize=20, fontweight='bold', color='darkblue')

# Show legend to explain the red line
plt.legend(loc = 'upper right')
plt.savefig('Charts/Facebook Stock Price Histogram.png', dpi=300, bbox_inches='tight')
# Display the plot
plt.show()


```

```python

import matplotlib.dates as mdates
# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Plot stock prices over time
df.plot(x='Date', y=['NFLX', 'FB', 'TWTR'], figsize=(18, 10), linewidth=3)

# Label axes
plt.ylabel('Price [$]')
plt.xlabel('Year')

# Format x-axis to show only years
ax = plt.gca()  # get current axis
ax.xaxis.set_major_locator(mdates.YearLocator())         # tick once per year
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y')) # format as YYYY

# Add title, legend, and grid
plt.title('Stock Prices Trend over Time', fontsize=20, fontweight='bold', color='darkblue')
plt.legend(loc='upper center')
plt.grid()

# Save before showing
plt.savefig('Charts/Stock Price Trend by Year.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Load Data
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
# Create a dataFrame named df_cancer with input/output data
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))

```

```python
# Task #8: Count plot (Seaborn)
plt.figure(figsize=(5,5))
# sns.countplot(df_cancer['target'], label = 'count')
sns.countplot(x='target', data=df_cancer)
plt.title('Distribution of Target Classes')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()
```
